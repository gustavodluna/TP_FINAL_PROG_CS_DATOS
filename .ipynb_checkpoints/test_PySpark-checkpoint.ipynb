```python
from pyspark.context import SparkContext
from pyspark.sql.session import SparkSession
sc=SparkContext.getOrCreate()
spark=SparkSession(sc)
```

```python
sc
```

```python
file="file.txt"
```

```python
lines=sc.textFile(file)
lines
```

```python
lines.count()
```

```python
lines.first()
```

```python
lines2=lines.sample(fraction=0.1,withReplacement=False)
```

```python
lines2.first()
```

```python
resultado=lines.filter(lambda line: "Samsung" in line)
```

```python
resultado
```

```python
resultado.count()
```

```python
resultado.take(1)
```

```python
enteros=lines.filter(lambda x: any(i.isdigit() for i in x))
enteros.take(5)
```

```python
enteros.count()
```

```python
enteros.persist()
```
