## • Dividimos el dataframe en dos partes, utilizando la columna "modelo" como PK

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

def dividir_df_con_pk(df, primary_key_col, n):
    # Obtener las columnas del DataFrame original
    columnas_originales = df.columns
    
    # Determinar la cantidad de columnas extra que se deben dividir equitativamente
    num_columnas_extras = len(columnas_originales) - 1
    
    # Calcular la cantidad de columnas extra que se asignarán a cada DataFrame
    columnas_por_df = num_columnas_extras // n
    
    # Lista para almacenar los DataFrames resultantes
    dfs = []
    
    # Iterar sobre la cantidad de divisiones
    for i in range(n):
        # Calcular el índice de inicio y fin para las columnas extras
        inicio_columnas_extras = i * columnas_por_df
        fin_columnas_extras = (i + 1) * columnas_por_df
        
        # Si es la última iteración, asegurarse de incluir todas las columnas extras restantes
        if i == n - 1:
            fin_columnas_extras = num_columnas_extras
        
        # Seleccionar las columnas para este DataFrame
        columnas_df = [primary_key_col] + \
                      [col for col in columnas_originales[1:] if col != primary_key_col][inicio_columnas_extras:fin_columnas_extras]
        
        # Seleccionar las columnas y agregar el DataFrame resultante a la lista
        dfs.append(df.select(columnas_df))
    
    return dfs
```

```python

# Llamar a la función para dividir el DataFrame en n partes
partes = dividir_df_con_pk(df, "modelo", 2)

# Mostrar los resultados
for i, parte in enumerate(partes):
    print(f"Parte {i+1}:")
    parte.show()
```

### • Utilizando acumuladores, determine la cantidad de teléfonos de la marca “Samsung” y “OnePlus”.

```python
# Convertir la columna 'modelo' en un RDD
modelo_rdd = df_joined.select('modelo').rdd.map(lambda x: x[0])
```

```python
modelo_rdd.take(10)
```

```python
def fun(x):
    #if x[0] in ["OnePlus","Samsung ","Xiaomi"]:
    if x.startswith(("Supported", "Not Supported")):
        return ((x,1))
    elif x.startswith("Hybrid"):
        return ((x,1))
    else:
        return ((x,1))
```

```python
def fun_r(x):
    #if x[0] in ["OnePlus","Samsung ","Xiaomi"]:
    if x.startswith("Samsung"):
        return ((x,4))
    elif x.startswith("OnePlus"):
        return ((x,3))
    elif x.startswith("Xiaomi"):
        return ((x,2))
    else:
        return ((x,1))
```

```python
op_ac=sc.accumulator(0)
ss_ac=sc.accumulator(0)
```

```python
def modelos(linea):
    global op_ac,ss_ac
ascii    if "OnePlus" in linea:
        op_ac+=1
        return True
    if "Samsung" in linea:
        ss_ac+=1
        return True
    else:
        return False
```

```python
map_func=modelo_rdd.filter(modelos)
```

### • Haciendo uso de MapReduce, determine la cantidad de teléfonos Samsung y cuantos de ellos son de la serie S.

```python
def modelos_map(x):
    if "Samsung" in x and "Galaxy S" in x:
        return ("Count",(1,1))
    elif "Samsung" in x:
        return ("Count",(1,0))
    elif "Galaxy S" in x:
        return ("Count",(0,1))
    else:
        return ("Count",(0,0))
```

```python
mapfun=modelo_rdd.map(modelos_map)
```

```python
mapfun.count()
```

```python
modelo_rdd.count()
```

```python
mapfun.reduceByKey(lambda x,y: (x[0]+y[0],x[1]+y[1])).collect()
```

```python
res=mapfun.reduceByKey(lambda x,y: (x[0]+y[0],x[1]+y[1])).collect()
```
